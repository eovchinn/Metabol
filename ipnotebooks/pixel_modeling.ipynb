```python
import os
from os import path
from datetime import datetime
import gensim
from gensim import corpora, models
import pandas as pd

import logging
logging.basicConfig(format='%(asctime)s : %(levelname)s : %(message)s', level=logging.INFO)

base_path = '/opt/data/pixel-corpora/'
ann_path = '/opt/data/pixel-annot-export/'

cor_path = path.join(base_path, 'pixel_corpus.mm')
tfidf_cor_path = path.join(base_path, 'pixel_tfidf.mm')
lsi_cor_path = path.join(base_path, 'pixel_lsi.mm')
lda_cor_path = None

numb_topics = 100

corpus = gensim.corpora.MmCorpus(cor_path)

ion_df_path = path.join(ann_path, 'ion_df.msgpack')
ion_df = pd.read_msgpack(ion_df_path)
dictionary = {idx : row['formula']+row['adduct'] for idx, row in ion_df.iterrows()}

if lda_cor_path:
    lda_model = gensim.models.ldamulticore.LdaMulticore(corpus, num_topics=numb_topics, id2word = dictionary, workers=4)
    lda_corpus = lda_model[corpus]
    corpora.MmCorpus.serialize(lda_cor_path, lda_corpus)

if tfidf_cor_path:
    tfidf_model = gensim.models.TfidfModel(corpus, id2word = dictionary)
    tfidf_corpus = tfidf_model[corpus]
    corpora.MmCorpus.serialize(tfidf_cor_path, tfidf_corpus)
    
    corpus = None
    tfidf_corpus = None
    tfidf_model = None

if lsi_cor_path:
    tfidf_corpus = gensim.corpora.MmCorpus(tfidf_cor_path)
    lsi_model = gensim.models.LsiModel(tfidf_corpus, id2word=dictionary, num_topics=numb_topics)
    lsi_corpus = lsi_model[tfidf_corpus]
    corpora.MmCorpus.serialize(lsi_cor_path, lsi_corpus)
    
    tfidf_corpus = None
    lsi_corpus = None
    lsi_model = None
```
