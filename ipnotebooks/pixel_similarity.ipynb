```python
import gensim
from gensim import similarities
from gensim.similarities import Similarity
from gensim import corpora
import os
from os import path

import logging
logging.basicConfig(format='%(asctime)s : %(levelname)s : %(message)s', level=logging.INFO)

# model param : tfidf or lsi or corpus
model = "tfidf"

base_path = '/opt/data/'
cor_path = path.join(base_path, 'pixel-corpora/pixel_'+model+'.mm')
sim_path = os.path.join(base_path,'pixel_similarity_models/'+model+'.index')
opref = os.path.join(base_path,'pixel_similarity_models/'+model+'sim')

corpus = gensim.corpora.MmCorpus(cor_path)

sim_index = gensim.similarities.docsim.Similarity(opref, corpus, num_features = corpus.num_terms)
sim_index.save(sim_path)
```
