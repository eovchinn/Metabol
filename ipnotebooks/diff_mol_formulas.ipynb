```python
file1 = '/home/katya/Dropbox/metabolite annotation/diff mol formulas/metaspace_annotations_testis 65um.csv'
file2 = '/home/katya/Dropbox/metabolite annotation/diff mol formulas/metaspace_annotations_7WB.csv'
ofile = None

import csv

def read_csv(file_path):
    res = {}
    with open(file_path, 'r') as csvfile:
        rowreader = csv.reader(csvfile, delimiter=',')
        first = True
        for row in rowreader:
            if first: first = False
            else:
                ds_name = row[1].replace(',',';')
                sf = row[2]
                mol_name = row[10].replace(',',';')
                n = sf+","+mol_name
                if not n in res: res[n] = []
                res[n].append(ds_name)
    return res

data1 = read_csv(file1)
data2 = read_csv(file2)

common_mols = set(data1.keys()) & set(data2.keys())

import sys
ofile = open(ofile, "w") if ofile else sys.stdout

for cm in common_mols:
    ofile.write(cm)
    for dn in data1[cm]+data2[cm]:
        ofile.write(';'+dn)
    ofile.write('\n')
```
