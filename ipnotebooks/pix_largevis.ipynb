```python
import numpy
import matplotlib.pyplot as plt

label_file = '/home/katya/Projects/Metabol/LargeVis/pix_label.lv'
lv_2d_file = '/home/katya/Projects/Metabol/LargeVis/pix_tfidf_2D.lv'
r = ''

label = []
for line in open(label_file):
    label.append(line.strip())
    
N = M = 0
all_data = {}
for i, line in enumerate(open(lv_2d_file)):
    vec = line.strip().split(' ')
    if i == 0:
        N = int(vec[0])
        M = int(vec[1])
    elif i <= N:
        all_data.setdefault(label[i-1], []).append((float(vec[-2]), float(vec[-1])))

colors = plt.cm.rainbow(numpy.linspace(0, 1, len(all_data)))

for color, ll in zip(colors, sorted(all_data.keys())):
    x = [t[0] for t in all_data[ll]]
    y = [t[1] for t in all_data[ll]]
    plt.plot(x, y, '.', color = color, markersize = 1)
if r != '':
    l = abs(float(r))
    plt.xlim(-l, l)
    plt.ylim(-l, l)
    
plt.show()
```
